```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

gpt4o_llm = ChatOpenAI(model="gpt-4o", temperature=0)
response = gpt4o_llm.invoke("Hello, how are you?")
response.pretty_print()
```

```python
gpt3_llm = ChatOpenAI(model="gpt-3.5-turbo-0125", temperature=0)
response = gpt3_llm.invoke("Hello, how are you?")
response.pretty_print()
```

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

msg_1 = SystemMessage(content="You are a helpful assistant.")
msg_2 = HumanMessage(content="Hello, my name is Fredy.")
msg_3 = AIMessage(content="Hello Fredy! How can I help you today?")
msg_4 = HumanMessage(content="What's my name?")
history = [msg_1, msg_2, msg_3, msg_4]
response = gpt3_llm.invoke(history)
response.pretty_print()
```

```python
from langchain_anthropic import ChatAnthropic

claude_llm = ChatAnthropic(model="claude-sonnet-4-20250514", temperature=0)
response = claude_llm.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("claude-sonnet-4-20250514", temperature=0)
response = llm.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("openai:gpt-4o", temperature=0)
response = llm.invoke(history)
response.pretty_print()
```
