```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model
llm = init_chat_model("openai:gpt-4o", temperature=0)
```

```python
file_search_tool = {
    "type": "file_search",
    "vector_store_ids": ["vs_691ba0bbfc0081918552d59f562187eb"],
}

llm_with_tools = llm.bind_tools([file_search_tool])
```

```python
response = llm_with_tools.invoke("What is the certification path for become a AWS Solutions Architect?")
response.pretty_print()
```
